```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import pearsonr
```

```python
#Import the data for model training and testing
df = pd.read_excel('Sentiment Analysis.xlsx')
df.head()
```

```python
#Using Pearson's Correlation, we can see the strength and direction of 2 variables.
dt = df.drop(columns= ['Date', 'HealthNeg', 'HealthPos', 'Health-RPD', 'EconNeg', 'EconPos', 'Econ-RPD', 'dJKSE', 'zJKSE'])
coef_corr = dt.corr()

sns.heatmap(data = coef_corr, annot=True)
```
